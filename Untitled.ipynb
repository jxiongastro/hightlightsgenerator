```python
from glob import glob
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline

plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['image.cmap'] = 'seismic'
plt.style.use('dark_background')
```

```python
def preview(f):
    plt.imshow(cv2.cvtColor(plt.imread(f), cv2.COLOR_RGB2GRAY))
```

```python
def preview_diff(f1, f2):
    img1 = cv2.cvtColor(plt.imread(f1), cv2.COLOR_RGB2GRAY) / 255.
    img2 = cv2.cvtColor(plt.imread(f2), cv2.COLOR_RGB2GRAY) / 255.
    diff = img1 - img2
    plt.imshow(diff)
    print('Mean:', np.mean(diff.flatten()))
    print('Median:', np.median(diff.flatten()))
    print('Std:', np.std(diff.flatten()))
    print()
    for p in [90,80,70,60,50,40,30,20,10]:
        print('{p}th Percentile:'.format(p=p), np.percentile(diff.flatten(), p))
    
preview_diff('frames/000000.jpg', 'frames/000025.jpg')
```

```python
def preview_abs_diff(f1, f2):
    img1 = cv2.cvtColor(plt.imread(f1), cv2.COLOR_RGB2GRAY) / 255.
    img2 = cv2.cvtColor(plt.imread(f2), cv2.COLOR_RGB2GRAY) / 255.
    diff = np.abs(img1 - img2)
    plt.imshow(diff)
    print('Mean:', np.mean(diff.flatten()))
    print('Median:', np.median(diff.flatten()))
    print('Std:', np.std(diff.flatten()))
    print()
    for p in [90,80,70,60,50,40,30,20,10]:
        print('{p}th Percentile:'.format(p=p), np.percentile(diff.flatten(), p))
    
preview_abs_diff('frames/000000.jpg', 'frames/000025.jpg')
```

```python
from skimage.measure import compare_ssim
from tqdm import tqdm
```

```python
frame_filenames = glob('frames/*.jpg')
N = len(frame_filenames)
```

```python
compare_ssim(
    cv2.cvtColor(plt.imread('frames/000000.jpg'), cv2.COLOR_RGB2GRAY), 
    cv2.cvtColor(plt.imread('frames/000025.jpg'), cv2.COLOR_RGB2GRAY))
```

```python
results = []
frame_filenames = np.array(sorted(frame_filenames))
for i in tqdm(range(N-1)):
    result = compare_ssim(
        cv2.cvtColor(plt.imread(frame_filenames[i]), cv2.COLOR_RGB2GRAY), 
        cv2.cvtColor(plt.imread(frame_filenames[i+1]), cv2.COLOR_RGB2GRAY))
    results.append(result)
results = np.array(results)
_ = plt.scatter(np.arange(len(results)), results, s=1)
```

```python
results = []
frame_filenames = np.array(sorted(frame_filenames))
for f in tqdm(frame_filenames):
    result = (cv2.cvtColor(plt.imread(f), cv2.COLOR_RGB2GRAY).flatten()/255.).mean()
    results.append(result)
results = np.array(results)
_ = plt.scatter(np.arange(len(results)), results, s=1)
```

```python
results = []
frame_filenames = np.array(sorted(frame_filenames))
for f in tqdm(frame_filenames):
    result = np.mean(cv2.Sobel(cv2.cvtColor(plt.imread(f), cv2.COLOR_RGB2GRAY), -1, 1, 1).flatten()/255.)
    results.append(result)
results = np.array(results)
_ = plt.scatter(np.arange(len(results)), results, s=1)
```

```python
results = []
frame_filenames = np.array(sorted(frame_filenames))
for f in tqdm(frame_filenames):
    result = np.mean(cv2.Laplacian(cv2.cvtColor(plt.imread(f), cv2.COLOR_RGB2GRAY), -1).flatten()/255.)
    results.append(result)
results = np.array(results)
_ = plt.scatter(np.arange(len(results)), results, s=1)
```

```python
results = []
frame_filenames = np.array(sorted(frame_filenames))
for f in tqdm(frame_filenames):
    result = cv2.Laplacian(cv2.cvtColor(plt.imread(f), cv2.COLOR_RGB2GRAY), -1).flatten()/255.
    result = result[result>0].mean()
    if result > 0.15:
        result = 0.15
    results.append(result)
results = np.array(results)
_ = plt.scatter(np.arange(len(results)), results, s=1)
```
